## Problem Statement 1

### IF P VALUE < 0.05, REJECT Ho

### IF P VALUE > 0.05, ACCEPT HO

1st Sample- Employee with Attrition as "Yes" with certain age values.

2nd Sample- Employee with Attrition as "No" with certain age values.

Both these samples are independent so we have to apply Mann Whitney Test

```python
from scipy.stats import mannwhitneyu
stats1,p1=mannwhitneyu(dataset3.Age,dataset4.Age)
print(stats1,p1)
```

H0: There is no significant differnce in the employee's age between Attrition(Yes) and Attrition(No).(Null hypothesis) 

H1: There is significant differnce in the employee's age between Attrition(Yes) and Attrition(No).(Alternate hypothesis) 


### The p value<0.05 so null hypothesis is rejected and alternate hypothesis is accepted

## Problem Statement 13

Both the vaiables are categorical so chi-squared test is used.

```python
from scipy.stats import chi2_contingency
chitable1=pd.crosstab(dataset2.Attrition,dataset2.BusinessTravel)
chitable1
```

```python
stats13,p13,dof,expeted=chi2_contingency(chitable1)
print(stats13,p13)
```

H0: Employee's business travel history does not have any significant effect over Attrition.(Null hypothesis)

H1: Employee's business travel history have significant effect over Attrition.(Alternate hypothesis)

### The p value<0.05 so null hypothesis is rejected and alternate hypothesis is accepted
